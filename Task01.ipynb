```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#2. random 100000x1 array
plt.hist(np.random.rand(100000), density=True, bins=50, histtype="step", color="blue", label="rand")
plt.hist(np.random.randn(100000), density=True, bins=50, histtype="step", color="red", label="randn")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc = "upper left")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
#3.
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Plot of Z = X² + Y²")
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
import seaborn as sns
import pandas as pd

!wget -q https://elitedatascience.com/wp-content/uploads/2022/07/Pokemon.csv
#reading dataset
df = pd.read_csv('Pokemon.csv', index_col=0, encoding='latin')

# assigning features
features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
data = df[features]

# Pearson
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(method='pearson'), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Pearson Correlation")
plt.show()
```

Pearson

```python
# Spearman
plt.figure(figsize=(8, 6))
sns.heatmap(data.corr(method='spearman'), annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Spearman Correlation")
plt.show()
```
